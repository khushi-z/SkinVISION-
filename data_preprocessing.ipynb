```python
# Mapped each image to its disease label
import os
import pandas as pd

import os

DATA_DIR = "data"
METADATA_PATH = os.path.join(DATA_DIR, "HAM10000_metadata.csv")
IMAGE_FOLDER = os.path.join(DATA_DIR, "HAM10000_images")

if not os.path.exists(METADATA_PATH):
    print(" Dataset not found.")
    print(" Please download HAM10000 dataset and place it inside 'data/' folder.")
else:
    df = pd.read_csv(METADATA_PATH)

    df["path"] = df["image_id"].apply(
        lambda x: os.path.join(IMAGE_FOLDER, f"{x}.jpg")
    )

    print(df.head())
```

```python
# verifying img mapping
import os
missing = df[~df["path"].apply(os.path.exists)]
print("Missing images:", len(missing))
```
